```python
def chunk(dString, n, step=None):
    if step is None: step = n
    return [dString[i:i+n] for i in range(0, len(dString)-n, step)]
```

```python
def toBin(s):
    output = ""
    for c in s:
        b = bin(ord(c))[2:]
        b = ((8-len(b)) * '0') + b
        output = output + b
    return output
```
